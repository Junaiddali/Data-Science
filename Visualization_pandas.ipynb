```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Ice Cream Ratings.csv")
df = df.set_index("Date")
df
```

```python
df.plot(kind = "line",subplots=True,title = "icream ratings", xlabel='daily ratings',ylabel='scores')
```

```python
df.plot(kind='bar',stacked = True)
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x='Texture Rating',y='Overall Rating',s=500, color='blue')
```

```python
df.plot.hist(bins=20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavor Rating',figsize=(10,10))
```
